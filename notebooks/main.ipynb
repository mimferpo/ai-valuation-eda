```python
analysis_base = pd.read_csv("../analysis_data/analysis_base.csv")
# displaying the number of rows and columns in the core_financials dataframe
core_financials.shape
```

```python
core_metrics = ["Total Revenue", "Net Income"]
metric_filter = analysis_base["metric"].isin(core_metrics)

core_financials = analysis_base.loc[metric_filter, :].copy()
core_financials = core_financials.sort_values(by=["symbol", "metric", "fiscal_year_end"])

core_financials.head()

```

```python
# latest revenue and net income
latest_core = (
    core_financials
    .sort_values(by=["symbol", "metric", "fiscal_year_end"])
    .groupby(["symbol", "metric"], as_index=False)
    .tail(1)
)

latest_core[["symbol", "metric", "fiscal_year_end", "amount"]]
```

```python
# valuation snapshot (one row per company)
valuation_snapshot = (
    analysis_base
    .drop_duplicates(subset=["symbol"])
    [["symbol", "price_to_sales", "trailingPE", "forwardPE", "profitMargins", "info_total_revenue"]]
    .set_index("symbol")
    .round(2)
)

valuation_snapshot
```

```python
# side-by-side comparison table
latest_wide = latest_core.pivot(
    index="symbol",
    columns="metric",
    values="amount",
).round(2)

comparison = valuation_snapshot.join(latest_wide)
```

```python
# year-over-year change 
yoy_change = (
    core_financials
    .sort_values(by=["symbol", "metric", "fiscal_year_end"])
    .groupby(["symbol", "metric"])["amount"]
    .pct_change()
    .mul(100)
    .round(2)
    .rename("yoy_pct_change")
)

core_financials.assign(yoy_pct_change=yoy_change)[
    ["symbol", "metric", "fiscal_year_end", "amount", "yoy_pct_change"]
]
```

```python
# valuation
comparison["valuation_label"] = np.where(
    comparison["price_to_sales"] == comparison["price_to_sales"].min(),
    "lower valuation",
    "higher valuation",
)

# margin
comparison["margin_label"] = np.where(
    comparison["profitMargins"] == comparison["profitMargins"].max(),
    "higher margin",
    "lower margin",
)

comparison
```
